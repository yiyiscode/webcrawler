```python
url = 'https://news.pts.org.tw/dailynews.php'
resp = requests.request(method='GET', url=url)
soup = BeautifulSoup(resp.text, features='html.parser')
nodes = soup.select('div > h2 > a')
titles = [node.text for node in nodes]
hrefs = [node.get('href') for node in nodes]
print(len(titles))
print(len(hrefs))
pd.DataFrame({'title':titles, 
              'href':hrefs})
```

```python
contents = []
summarys = []
for href in hrefs:
    resp = requests.request(method='get', url=href)
    soup = BeautifulSoup(resp.text, features='html.parser')
    nodes = soup.select('div.post-article > p') #get class like list
    summarys.append(soup.select('div.post-article > div')[0].text)
    contents.append(''.join([node.text for node in nodes]))

print(len(contents))
```

```python
page = 2
url = f'https://news.ltn.com.tw/ajax/breakingnews/all/{page}'
resp = requests.request(method='get', url=url)
resp_json = json.loads(resp.text)
# print(type(resp_json))
# print(resp_json.keys())
if page == 1:
    data_lst = resp_json.get('data')#list contents dict
    titles = [data.get('title') for data in data_lst]
    urls = [data.get('url') for data in data_lst]
    summarys = [data.get('summary').replace('\n', '') for data in data_lst]
else:
    start = page*10
    end = start+20
    data_dict = resp_json.get('data')#dict
    titles = [data_dict.get(str(i)).get('title') for i in range(start, end)]
    urls = [data_dict.get(str(i)).get('url') for i in range(start, end)]
    summarys = [data_dict.get(str(i)).get('summary') for i in range(start, end)]
```

```python
contents = []
for url in urls:
    resp = requests.get(url)
    soup = BeautifulSoup(resp.text)
    nodes = soup.select('div.text > p')
    contents.append(''.join([node.text for node in nodes[1:-2]]))
```

### <font color=#26734d>Connect to mysql function</font>

```python
def connect_mysql(data, user, password, host, dbname, tablename, if_exists='fail'):
    from sqlalchemy import create_engine
    import pymysql
    # cretae engine
    engine = create_engine(f'mysql+pymysql://{user}:{password}@{host}:3306/{dbname}')
    # "sqltype+moudlename://user:password@host:port/dbname"

    # create connection
    conn = engine.connect()
    try:
        # add timestamp
        now = time.strftime("%Y/%m/%d %H:%M:%S")
        data['time'] = now
        data.to_sql(name=tablename, con=conn, 
                if_exists=if_exists, index=False)
    except ValueError as vx:
        print(vx)
    except Exception as ex:
        print(ex)
    else:
        print("Table {} {}s in {} sccessfully.".format(tablename, if_exists, dbname))
    finally:
        conn.close()
```

```python
tablename = 'ltn_breakingnews'
connect_mysql(data, user, password, host, dbname, tablename, if_exists='replace')
```

```python
class Stock:
    def __init__(self):
        self._url = ''
        self._stockid_resp = {}
        
    def get_Concection(self, stockid):
        self._url = f'https://goodinfo.tw/tw/StockDetail.asp?STOCK_ID={stockid}'
        payload = {'STOCK_ID': stockid}
        headers = {'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/104.0.0.0 Safari/537.36'}
        self._resp = requests.request(method='POST', url=self._url, data=payload, headers=headers)
        if self._resp.ok:
            self._resp.encoding = 'utf-8'
            self._stockid_resp.update({stockid: self._resp})
            # self._stockid_resp[stockid] = self._resp (slower)
            return self._stockid_resp
        else:
            print(f'Error({resp.status_code}):{resp.reason}')
        if resp.status_code==404:
            return "Please check your url !!!"
        elif resp.status_code==403:
            return "Please give full headers !!!"
        else:
            print("https://developer.mozilla.org/en-US/docs/Web/HTTP/Status")
            return "Mayebe you should try other ways like selenium !!!"
    
    def get_Info(self, stockid):
        data = pd.read_html(self._stockid_resp.get(stockid).text)
        data = data[41]
        data.columns = data.iloc[0]
        data = data.iloc[1:]
        return data
               
    def show_AllData(self, stockid):
        data = pd.read_html(self._stockid_resp.get(stockid).text)
        for idx,  d in enumerate(data, start=0):
            print('-'*15, idx, '-'*15)
            display(d)
    
    def get_Data(self, stockid, idx):
        data = pd.read_html(self._stockid_resp.get(stockid).text)
        data[idx].insert(loc=0, column='股票號碼',
                        value=stockid, 
                        allow_duplicates=False)
        return data[idx]
```

```python
url = 'https://www.ptt.cc/ask/over18'
# set cookie -> over18=1
headers = {'Set-Cookie':'over18=1;'}
# set f'/bbs/Gossiping/index{page}.html' 
# to change page
payload = {'from': '/bbs/Gossiping/index.html',
           'yes': 'yes'}
resp = requests.post(url, data=payload,
                     headers=headers)
resp.encoding ='utf-8'
soup = BeautifulSoup(resp.text, 'html.parser')
nodes = soup.select('div.title > a')
hrefs = [node.get('href') for node in nodes]
titles = [node.text for node in nodes]
```

```python
conts = []
posttimes = []
authors =[]
for href in hrefs:
    url = 'https://www.ptt.cc/ask/over18'
    headers = {'Set-Cookie':'over18=1;'}
    payload = {'from': href,
               'yes': 'yes'}
    resp = requests.post(url, data=payload,
                         headers=headers)
    soup = BeautifulSoup(resp.text, 'html.parser')
    posttime = soup.select('div:nth-child(4) > span.article-meta-value')[0].text
    posttimes.append(posttime)
    author = soup.select('#main-content > div:nth-child(1) > span.article-meta-value')[0].text
    authors.append(author)
    conts.append(soup.select('#main-content')[0].text)
    time.sleep(1)
```

```python
data = pd.DataFrame({'title':titles,
                     'author':authors,
                     'posttime':posttimes,
                     'url':hrefs,
                     'conts':conts})
data.head(3)
```
